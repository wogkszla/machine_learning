```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# set random state to make experiment repeatable
rs = np.random.RandomState(1234)

# now generate 100 samples of X randomly from a normal distribution with mean = 5 (loc). 
# Many variables in real life follow this distribution.
X = rs.normal(loc=5,scale=5,size=100)

# For those who doubt here is a histogram of X :).
g = sns.distplot(X, kde=False, rug=True)
plt.axvline(X.mean(), color='r', linestyle='--')

plt.title("Histogram of X")
plt.xlabel("X")
plt.show(g)
```

```python
# Y positively correlated with X. means Y generally increases as X increases.
Y_positive_correlation = X + 2

# Y negatively correlated with X. means Y generally decreases as X increases
Y_negative_correlation = 20 - X
```

```python
import pandas as pd

pos_corr_pd = pd.DataFrame({
    'X': X,
    'Y_positive_correlation': Y_positive_correlation
})

neg_corr_pd = pd.DataFrame({
    'X': X,
    'Y_negative_correlation': Y_negative_correlation
})
```

```python
# the positive correlation plot.
g = sns.lmplot(x="X",y="Y_positive_correlation", data=pos_corr_pd, line_kws=dict(color="r") )
plt.title('Positive Correlation (r = 1)')
plt.show(g)

# the negative correlation plot
g = sns.lmplot(x="X",y="Y_negative_correlation", data=neg_corr_pd, line_kws=dict(color="r") )
plt.title('Negative Correlation (r = -1)')
plt.show(g)
```

```python
np.corrcoef(X,Y_positive_correlation), np.corrcoef(X,Y_negative_correlation)
```

```python
# Let's create and visualise different correlations
import math as m
import numpy as np
rs = np.random.RandomState(1234)

noise = rs.normal(loc=2,scale=3,size=100)
X = rs.normal(loc=5,scale=5,size=100)

# X is directly related to Y with a bit of noise
Y_strong_positive = X + noise

# Noisy negative correlation. Y decreases with increasing X with some noise
Y_strong_negative = 20 - X + noise

# Y NOT related to X
Y_not_related = noise

# Y is related to X differently (sinX)
Y_sin_function = [m.sin(x) for x in X]
```

```python
Y_strong_positive_df = pd.DataFrame({
    'X': X,
    'Y_strong_positive': Y_strong_positive
})

Y_strong_negative_df = pd.DataFrame({
    'X': X,
    'Y_strong_negative': Y_strong_negative
})

Y_not_related_df = pd.DataFrame({
    'X': X,
    'Y_not_related': Y_not_related
})

Y_sin_function_df = pd.DataFrame({
    'X': X,
    'Y_sin_function': Y_sin_function
})
```

```python
g = sns.lmplot(x="X",y="Y_strong_positive", data=Y_strong_positive_df, line_kws=dict(color="r") )
plt.title('Positive Correlation (r = 1)')
plt.show(g)

g = sns.lmplot(x="X",y="Y_strong_negative", data=Y_strong_negative_df, line_kws=dict(color="r") )
plt.title('Negative Correlation (r = 1)')
plt.show(g)

g = sns.lmplot(x="X",y="Y_not_related", data=Y_not_related_df, line_kws=dict(color="r") )
plt.title('Not Related Correlation (r = 1)')
plt.show(g)

g = sns.lmplot(x="X",y="Y_sin_function", data=Y_sin_function_df, line_kws=dict(color="r") )
plt.title('Sin Function Correlation (r = 1)')
plt.show(g)
```
